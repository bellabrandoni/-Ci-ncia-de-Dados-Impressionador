# Variáveis

## Escalas de mensuração

As variáveis ainda podem ser classificadas de acordo com o nível ou escala de mensuração. **Mensuração** é o processo de atribuir números ou rótulos a objetos, pessoas, estados ou eventos de acordo com regras específicas para representar quantidades ou qualidades dos atributos.

*Regra* é um guia, método ou comando que indica ao investigador como medir o atributo.

*Escala* é um conjunto de símbolos ou números, construído com base em uma regra, e aplica-se a indivíduos ou a seus comportamentos ou atitudes. A posição de um indivíduo na escala é baseada na posse dele do atributo que a escala deve medir.

## Número de categorias e escalas de precisão

### Variáveis categóricas

As variáveis categóricas podem ser classificadas de acordo com o número de categorias em:

- dicotômicas (binárias, dummy): possuem apenas duas categorias, como sim/não, verdadeiro/falso, 0/1, etc. Exemplos: fumantes(1)/não fumantes(0), doentes(1)/não doentes(0), etc.
- policotômicas: possuem mais de duas categorias. Exemplos: cor dos olhos (azul, verde, castanho, ...), grau de escolaridade (fundamental, médio, superior, ...), etc.

### Variáveis quantitativas

Quanto a escala de precisão, as variáveis quantitativas podem ser classificadas em:

- discretas: assumem valores inteiros e finitos. Exemplos: número de filhos, número de carros, número de acidentes, etc.
- contínuas: assumem valores reais e infinitos. Exemplos: peso, altura, temperatura, salário, etc.
