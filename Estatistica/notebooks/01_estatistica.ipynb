# Estatística

A estatística pode ser definida como a ciência que tem por objetivo a coleta, análise e interpretação de dados qualitativos e quantitativos.

Está dividida em três grandes áreas: estatística descritiva ou dedutiva, estatística probabilística e estatística inferencial ou indutiva.

O objetivo da *estatística descritiva* é obter uma descrição do comportamento da variável em estudo por meio de um tratamento adequado dos dados.

A *estatística probabilística* é a parte da estatística que estuda os fenômenos aleatórios, ou seja, aqueles que não podem ser previstos com certeza. A probabilidade é a medida da incerteza associada a um evento aleatório.

Com a *estatística inferencial* é possível fazer inferências sobre uma população a partir de uma amostra. A inferência estatística é um conjunto de técnicas que permitem generalizar as conclusões obtidas a partir de uma amostra para a população. Com métodos inferenciais também é possível analisar a representatividade dos resultados, a significância e a confiabilidade dos estudos realizados.

Dentre os elementos básicos da estatística, podemos citar:

- população (ou universo): conjunto que contém todos os indivíduos, objetos ou elementos a serem estudados, que apresentam uma ou mais características em comum. Por exemplo, a população de um estudo sobre a altura dos alunos de uma escola é o conjunto de todos os alunos da escola.
- amostra: subconjunto da população, selecionado para ser estudado. A amostra deve ser representativa da população, ou seja, deve conter as mesmas características da população. No exemplo anterior, a amostra seria o grupo de alunos selecionados para participar do estudo sobre a altura. O processo de escolha de uma amostra da população é chamado de amostragem.
- variável: característica ou atributo que se deseja observar, medir ou contar, a fim de se obter algum tipo de conclusão. Por exemplo, a altura dos alunos citados anteriormente é uma variável. Uma empresa listada na bolsa de valores pode ter como variáveis o preço das ações, o lucro líquido, o número de funcionários, o setor de atuação, entre outras.
- dados: registros coletados sobre as variáveis de interesse.

## Tipos de variáveis

Variável é uma característica da população (ou amostra) em estudo, possível de ser medida, contada ou categorizada.

Exemplos de variáveis qualitativas (ou categóricas) são: sexo, cor dos olhos, estado civil, grau de instrução, sabor de sorvete preferido, entre outras. Essas variáveis não podem ser medidas, mas sim categorizadas.

Exemplos de variáveis quantitativas (ou numéricas) são: idade, altura, peso, temperatura, salário, número de filhos, entre outras. Essas variáveis podem ser medidas e expressas em números.
